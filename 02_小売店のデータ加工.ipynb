## データの確認

```python
# openpyxlをインストールしておく
uriage = pd.read_csv('./assets/2章/uriage.csv')
kokyaku = pd.read_excel('./assets/2章/kokyaku_daicho.xlsx')
# データ確認
display("売上データ",uriage.head())
display("顧客データ",kokyaku.head())
```

## データ補正せずに集計

```python
# item_nameの表記揺れ、item_priceの欠損など、データが綺麗でない。
# 売上日をデート型に変換
uriage["purchase_date"] = pd.to_datetime(uriage["purchase_date"])
# 月の売上カラムを追加
uriage["purchase_month"] = uriage["purchase_date"].dt.strftime("%Y%m")
uriage.head()
```

```python
# 月-商品ごとの売上を集計する
#res = uriage.pivot_table(index='purchase_month', columns='item_name', values='item_price', aggfunc='sum', full_value=0)
res = uriage.pivot_table(index="purchase_date", columns="item_name", values="item_price", aggfunc="sum")
res
```

```python
# 表記揺れがあるため、カラムの量が増えている。sumもうまく機能していない
# 04 商品名の表記揺れを修正する
# 全て大文字に変換する
uriage["item_name"] = uriage["item_name"].str.upper()
uriage.head()
```

```python
# 半角スペースを空文字に変換する
uriage["item_name"] = uriage["item_name"].str.replace(" ", "")
display(uriage.head())
print(pd.unique(uriage.item_name))
print(len(pd.unique(uriage.item_name)))
```

```python
# A-Zの26個の商品に修正できてそう
# 05 item_priceの欠損値を補完する
# 欠損値の有無を確認
print(uriage.isnull().any())
```

```python
# 欠損情報を格納する
isnull = uriage["item_price"].isnull()
# 欠損値が含まれる商品名を抽出する
null_item_name = list(uriage.loc[isnull, "item_name"].unique())
null_item_name
```

```python
# 欠損値が含まれる商品名で処理を回す
for item_name in null_item_name:
    # 同じ商品で正しく設定されている価格を抽出し、その最大値を保持する
    reprice = uriage.loc[(uriage["item_name"] == item_name) & (isnull == False), "item_price"].max()
    # 同じ商品でitem_priceがNaNデータをrepriceに設定する
    uriage["item_price"].loc[(uriage["item_name"] == item_name) & (isnull == True)] = reprice

# 修正後のテーブルを確認   
display(uriage.head())
# 欠損値の有無を検算
display(uriage.isnull().any())
```

## 16 顧客名の揺れを補正

```python
kokyaku_data = pd.read_excel('./assets/2章/kokyaku_daicho.xlsx')
# データの確認
kokyaku_data.head()
```

```python
# 全角/半角スペースが入っていたり、統一されていない
# スペースの削除 
kokyaku_data["顧客名"] = kokyaku_data["顧客名"].str.replace("　", "")
kokyaku_data["顧客名"] = kokyaku_data["顧客名"].str.replace(" ", "")
kokyaku_data.head()
```

## 17 日付の揺れを補正

```python
# 登録日の書式が統一されていない
#  数値であるかどうかを判定し、配列に格納
fig_is_serial = kokyaku_data["登録日"].astype("str").str.isdigit()
print(fig_is_serial)
# 数値になっているデータ数をカウント
fig_is_serial.sum()
```

```python
# 数値になっているものを参照し、タイムスタンプに変換する
fromSerial = pd.to_timedelta(kokyaku_data.loc[fig_is_serial, "登録日"].astype("float"), unit="D") + pd.to_datetime('1900/01/01')
fromSerial
```

```python
# 書式を統一するため、その他のデータも処理する
fromString = pd.to_datetime(kokyaku_data.loc[~fig_is_serial, "登録日"])
fromString
```

```python
# 補正データを結合し、元データに代入する
kokyaku_data["登録日"] = pd.concat([fromString, fromSerial])
kokyaku_data.head()
```

```python
# 登録年月を集計する
kokyaku_data["登録年月"] = kokyaku_data["登録日"].dt.strftime("%Y-%m")
display(kokyaku_data.groupby("登録年月").count())
```

## 顧客名をキーに2つのデータをジョインする
